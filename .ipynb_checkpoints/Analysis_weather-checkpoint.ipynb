```python
import pandas as pd
import numpy as np
import os
```

```python
os.listdir('prep/data')
```

```python
def drop_unnamed(data):
    names = data.columns
    for name in names:
        if "Unnamed" in name:
            data.pop(name)
    return data
```

```python
pf = pd.read_csv("prep/data/new_performance_ver4.csv")
pf = drop_unnamed(pf)
pf.head(2)
```
