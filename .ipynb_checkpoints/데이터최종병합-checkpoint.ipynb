```python
import pandas as pd
import numpy as np
```

```python
pf = pd.read_csv('new_performance_ver4.csv', index_col = 0, parse_dates=['방송일시', 'date'])
pf.tail()
```

```python
pf['hour'] = pf['방송일시'].dt.hour
pf.head()
```

```python
wt = pd.read_csv('월별인구가중합날씨.csv', parse_dates = ['time'])
wt.head()
```

```python
wt['date'] = pd.to_datetime(wt['time'].dt.date)
wt['hour'] = wt['time'].dt.hour
wt.head()
```

```python
wt.dtypes
```

```python
del wt['time']
```

```python
npf = pd.merge(pf, wt, on=['date', 'hour'])
npf.tail()
```

```python
pf_p =pd.read_csv("performance_prime.csv" , index_col = 0, parse_dates = ['방송일시', 'date'])
pf_p.tail()
```

```python
pf_p.isnull().sum()
```

```python
pf2 = pd.concat([npf, pf_p], axis=1)
pf2.head()
```
