```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("final_performance_v4.csv", parse_dates=['방송일시','date', 'real_date'])
data.head()
```

```python
use_var = ['마더코드', '상품코드', '상품군', 'week', 'hour', 'weekofyear', 'stage', 'prime', 'IsHoliday', '노출(분)', '판매단가', '취급액', '지속휴일수', 'TEMP', 'HUM', 'cpi']
Data = data[use_var]
Data.head()
```

```python
Data['profit/m'] = Data.apply(lambda x: x['취급액']/x['노출(분)'], axis=1)
Data.head()
```

```python
avg = Data[['상품코드', 'profit/m']].groupby('상품코드').mean()
avg.rename(columns = {'profit/m': 'avgp'}, inplace=True)
avg
```

```python
minval = Data[['상품코드', 'profit/m']].groupby('상품코드').min()
minval.rename(columns = {'profit/m': 'minp'}, inplace=True)
minval
```

```python
maxval = Data[['상품코드', 'profit/m']].groupby('상품코드').max()
maxval.rename(columns={'profit/m': 'maxp'}, inplace=True)
maxval
```

```python
Data = pd.merge(Data, avg, on=['상품코드'])
Data
```

```python
Data = pd.merge(Data, minval, on=['상품코드'])
Data.head()
```

```python
Data = pd.merge(Data, maxval, on=['상품코드'])
Data.head()
```

```python
Data.to_csv("final_performance_with_stat.csv", mode='w')
```

```python
import sklearn
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
use_var = ['마더코드', '상품코드', '상품군', 'week', 'hour', 'weekofyear', 'stage', 'prime', 'IsHoliday', '노출(분)', '판매단가', '취급액', '지속휴일수', 'TEMP', 'HUM', 'cpi', 'avgp', 'minp', 'maxp']
Data = Data[use_var]
Data.head()
```

```python
Data['avgp'] = Data.apply(lambda x: x['노출(분)']*x['avgp'], axis=1)
Data['minp'] = Data.apply(lambda x: x['노출(분)']*x['minp'], axis=1)
Data['maxp'] = Data.apply(lambda x: x['노출(분)']*x['maxp'], axis=1)
Data.head()
```

```python
Data['상품군'].unique()
```

```python
X = Data.drop(['취급액'], axis = 1)
Y = Data[['취급액']]
print("data shape : ", X.shape)
print("label shape : " , Y.shape)
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(X, Y, test_size = 0.1, random_state = 123457)
print("x_train shape : ", x_train.shape)
print("y_train shape : ", y_train.shape)
print("x_valid shape : ", x_valid.shape)
print("y_valid shape : ", y_valid.shape)
```

## Make valid data

```python
pd.set_option('mode.chained_assignment',  None)
def goods_code_encoding(row, code_set):
    if row['상품코드'] in code_set:
        return row
    else:
        if row['마더코드'] in x_train['마더코드']:
            frame = x_train[x_train['마더코드']==row['마더코드']]
            frame['diff'] =  abs(frame['판매단가'] - row['판매단가'])
            fitrow = frame[frame['diff']==min(frame['diff'])].iloc[0, :-1]
        else:
            frame = x_train[x_train['상품군']==row['상품군']]
            frame['diff'] =  abs(frame['판매단가'] - row['판매단가'])
            fitrow = frame[frame['diff']==min(frame['diff'])].iloc[0, :-1]
        row['avgp'] = fitrow['avgp']
        row['minp'] = fitrow['minp']
        row['maxp'] = fitrow['maxp']
        return row
```

```python
make_valid_cols = ['상품군', 'week', 'hour', 'weekofyear', 'stage', 'prime', 'IsHoliday', '노출(분)', '판매단가', '지속휴일수', 'TEMP', 'HUM', 'cpi', 'avgp', 'minp', 'maxp']
to_cat = ['상품군', 'week', 'hour', 'weekofyear', 'stage']
to_int = ['상품군', 'week', 'hour', 'weekofyear', 'stage', 'prime', 'IsHoliday']

train_goods_set = set(x_train['상품코드'])

for i in range(0, len(x_valid)):
    row = x_valid.iloc[i]
    x_valid.iloc[i] = goods_code_encoding(row, train_goods_set)

#for col in to_cat:
#    x_valid[col] = x_valid[col].astype('category') 
    
for col in to_int:
    x_valid[col] = x_valid[col].astype('int')
    
x_valid_input = x_valid[make_valid_cols]  
print("x_valid_input shape : ", x_valid_input.shape)
```

## Make train data

```python
make_train_cols = ['상품군', 'week', 'hour', 'weekofyear', 'stage', 'prime', 'IsHoliday', '노출(분)', '판매단가', '지속휴일수', 'TEMP', 'HUM', 'cpi', 'avgp', 'minp', 'maxp']
to_cat = ['상품군', 'week', 'hour', 'weekofyear', 'stage']
to_int = ['상품군', 'week', 'hour', 'weekofyear', 'stage', 'prime', 'IsHoliday']

#for col in to_cat:
#    x_train[col] = x_train[col].astype('category') 
for col in to_int:
    x_train[col] = x_train[col].astype('int')

x_train_input = x_train[make_train_cols]
print("x_train_input shape : ", x_train_input.shape)
```

## lightgbm 데이터셋으로 변환

```python
import lightgbm as lgb

train_ds = lgb.Dataset(x_train_input, label = y_train) 
valid_ds = lgb.Dataset(x_valid_input, label = y_valid)
```

## 파라미터 세팅

```python
params = {'learning_rate': 0.1, 
          'max_depth': -1, 
          'boosting': 'dart', 
          'objective': 'regression', 
          'metric': 'mape', 
          'is_training_metric': True, 
          'num_leaves': 144, 
          'feature_fraction': 0.9, 
          'bagging_fraction': 0.7, 
          'bagging_freq': 5, 
          'seed':123457}
```

## 모델훈련

```python
model = lgb.train(params, train_ds, 2000, valid_ds, verbose_eval=100, early_stopping_rounds=10)
```

## 모델평가

```python
predict_train = model.predict(x_train_input).reshape(-1, 1)
predict_test = model.predict(x_valid_input).reshape(-1, 1)
```

```python
import numpy as np

def MAPE(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
train_mape = MAPE(y_train, predict_train)
test_mape = MAPE(y_valid, predict_test)

print('train data MAPE: ', train_mape)
print('test data MAPE: ', test_mape)
```

# catboost

```python
import numpy as np
from catboost import Pool, CatBoostRegressor

to_cat = ['상품군', 'week', 'hour', 'weekofyear', 'stage','prime', 'IsHoliday']

train_pool = Pool(x_train_input, y_train, cat_features=to_cat)
test_pool = Pool(x_valid_input, cat_features=to_cat)

model = CatBoostRegressor(iterations=10000, loss_function='MAPE', verbose=100, early_stopping_rounds=100)

model.fit(train_pool)
```

```python
preds = model.predict(test_pool).reshape(-1,1)
valid_mape = MAPE(y_valid, preds)
print('test data MAPE: ', valid_mape)
```
