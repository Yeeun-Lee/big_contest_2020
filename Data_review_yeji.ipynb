```python
import pandas as pd
import numpy as np
```

```python
performance = pd.read_excel('performance.xlsx', header=1 )
performance
```

```python
performance['year'] = performance['방송일시'].map(lambda x: x.year)
performance['month'] = performance['방송일시'].map(lambda x: x.month)
performance['day'] = performance['방송일시'].map(lambda x: x.day)
performance['time'] = performance['방송일시'].map(lambda x: x.time())

performance
```
