```python
def goods_code_encoding(x_train, row, code_set):
#     print(row)
    if row['상품코드'] in code_set:
        return row
    else:
        if row['마더코드'] in x_train['마더코드']:
            frame = x_train[x_train['마더코드']==row['마더코드']]
            print(frame.head())
            print(frame.info())
            frame['diff'] =  abs(frame['판매단가'] - row['판매단가'])
            fitrow = frame[frame['diff']==min(frame['diff'])].iloc[0, :-1]

        else:
            frame = x_train[x_train['상품군']==row['상품군']]
            print(frame.head())
            print(frame.info())
            frame['diff'] =  abs(frame['판매단가'] - row['판매단가'])
            fitrow = frame[frame['diff']==min(frame['diff'])].iloc[0, :-1]
        row['avgp'] = fitrow['avgp']
        row['minp'] = fitrow['minp']
        row['maxp'] = fitrow['maxp']

        return row
```

```python
_X = test.drop(['방송일시', '상품명', '취급액', 'date',
                'time', 'real_date', 'profit/m'], axis = 1)
print(_X.columns)
_Y = test['취급액']
x_train, x_valid, y_train, y_valid = train_test_split(_X, _Y,
                                                      test_size=0.1,
                                                      random_state=123457)
# # filtering logic
# x_valid = filtering(x_train, x_valid)
train_goods_set = set(x_train['상품코드'])
for i in range(len(x_valid)):
    row = x_valid.iloc[i]
    x_valid.iloc[i] = goods_code_encoding(x_train, row, train_goods_set)
```

```python
pf = pd.read_excel("data/02_평가데이터/2020 빅콘테스트 데이터분석분야-챔피언리그_2020년 6월 판매실적예측데이터(평가데이터).xlsx",
                           sheet_name='6월편성', skiprows=1)
pf
```

```python
def caldate(date, time):
    """
    week : 0 - 월, 1 - 화, 2 - 수 ...
    """
    start = pd.Timestamp('00:00:00').time()
    end = pd.Timestamp('02:00:00').time()
    if time >= start and time <= end:
        date = date - pd.DateOffset(days = 1)
    return date
def stage(x):
    if x < 0.34:
        return 'early'
    elif 0.34 <= x < 0.67:
        return 'mid'
    return 'late'

def stage_advanced(x):
    if x['count'] == 1:
        return 'short'
    if x['count'] == 2:
        if x['p'] <= 0.5:
            return 'early'
        else:
            return 'late'
    else:
        return stage(x['p'])
```

```python
# 방송일시 처리
print("----방송일시 처리----")
pf['real_date'] = pd.to_datetime(pf['방송일시'].dt.date)
pf['week'] = pf['방송일시'].dt.dayofweek
pf['weekofyear'] = pf['방송일시'].dt.week
pf['year'] = pf['방송일시'].dt.year
pf['month'] = pf['방송일시'].dt.month
pf['time'] = pf['방송일시'].dt.time
pf['hour'] = pf['방송일시'].dt.hour.astype('U')

pf['date'] = pf.apply(lambda x: caldate(x['real_date'], x['time']), axis=1)
pf['week'] = pf['date'].dt.dayofweek
```

```python
# add Stage column
print("----방송 초/중/후반 정보 삽입----")
sample = pf[['방송일시', 'date', '상품코드', '노출(분)']]
sample['cumcount'] = sample.groupby(['date', '상품코드'])['노출(분)'].cumcount() + 1
sample2 = sample.groupby(['date', '상품코드'])['노출(분)'].count().reset_index()
sample2.rename({'노출(분)': 'count'}, axis=1, inplace=True)
sample = sample.merge(sample2, on=['date', '상품코드'])
sample['p'] = sample['cumcount'] / sample['count']

sample['stage'] = sample.apply(stage_advanced, axis=1)
pf = pf.merge(sample[['방송일시','상품코드', 'stage']], on = ['방송일시', '상품코드'])
```
