### 측정망 중 '도시대기'를 기준으로 전처리를 진행함 - sample(2019.07)

```python
data = sample[sample['망'] == '도시대기']
print(data.info())
print(data['지역'].unique())
data.tail()
```

```python
from datetime import datetime
```

```python
def convert_time(mytime):
    mytime = str(mytime)
    y = int(mytime[:4])
    m = int(mytime[4:6])
    d = int(mytime[6:8])
    H = int(mytime[8:10]) - 1
    mytime = datetime(y, m, d, H)
    return mytime
```

```python
data['측정일시'].astype('U').apply(lambda x: convert_time(x))
# .apply(lambda x: x.strftime("%Y-%m-%d"))
```

```python
data['지역'] = data['지역'].apply(lambda x: x.split()[0])
data['date'] = data['측정일시'].astype('U').apply(lambda x: convert_time(x)).apply(lambda x: x.strftime("%Y-%m-%d"))
data['month'] = data['측정일시'].astype('U').apply(lambda x: convert_time(x)).apply(lambda x: x.strftime("%Y-%m"))
data['hour'] = data['측정일시'].astype('U').apply(lambda x: convert_time(x)).apply(lambda x: x.strftime("%H"))
data = data.drop(['망', '측정소코드', '측정소명', '측정일시', '주소'], axis = 1)

data.head()
# df = data.groupby('지역').mean().reset_index()
```

```python
data['지역'].unique()
```

```python
data1 = data.groupby(['date', 'month', 'hour','지역']).mean().reset_index()
```

```python
pop = pd.read_csv("population_proportion.csv")
pop.index = pd.date_range('2019-01', '2020-08', freq='M').strftime('%Y-%m')
del pop['전국']
pop
```

```python
pop1 = pop.unstack().reset_index()
pop1.columns = ['region', 'month', 'p']
pop1.head()
```

```python
def change_region(x):
    regions = {'서울' : '서울특별시',
              '인천' : '인천광역시',
              '부산' : '부산광역시',
              '대구' : '대구광역시',
              '광주' : '광주광역시',
              '대전' : '대전광역시',
              '울산' : '울산광역시',
              '경기' : '경기광역시',
              '강원' : '강원도',
              '충북' : '충청북도',
              '충남' : '충청남도',
              '전북' : '전라북도',
              '전남' : '전라남도',
              '경북' : '경상북도',
              '경남' : '경상남도',
              '제주' : '제주도',
              '세종' : '세종특별자치시'}
    for key, value in regions.items():
        if x == key:
            return value
```

```python
data1['region'] = data1['지역'].apply(change_region)
del data1['지역']
data1.head()
```

```python
df = data1.merge(pop1, on = ['region', 'month'])
df.head()
```

```python
def mul(x):
    return x*df['p']
```

```python
df[['SO2', 'CO', 'NO2', 'PM10', 'PM25']] = df[['SO2', 'CO', 'NO2', 'PM10', 'PM25']].apply(lambda x: mul(x)).values
```

```python
del df['p']
df = df.groupby(['date', 'month', 'hour']).sum().reset_index()
df.head()
```

```python
del df['month']
```

```python
df.to_csv("미세먼지201901.csv", index = False)
```
