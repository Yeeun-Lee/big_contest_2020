```python
import pandas as pd
import numpy as np
```

```python
wt = pd.read_csv('월별인구가중합날씨_1.csv', parse_dates = ['time'])
wt.head()
```

```python
wt[8732: 8750]
```

```python
wt['date'] = pd.to_datetime(wt['time'].dt.date)
wt['hour'] = wt['time'].dt.hour
wt.head()
```

```python
wt.dtypes
```

```python
del wt['time']
```

```python
wt.rename(columns={'date':'real_date'}, inplace=True)
wt
```

```python
pf_p =pd.read_csv("performance_prime.csv" , index_col = 0, parse_dates = ['방송일시', 'date'])
pf_p = pf_p.reset_index(drop=True)
pf_p.tail()
```

```python
pf_p['real_date'] = pd.to_datetime(pf_p['방송일시'].dt.date)
pf_p.head()
```

```python
len(pf_p)
```

```python
pf_p.isnull().sum()
```

```python
holi = pd.read_csv("2019_2020년도_지속_휴일_수.csv", index_col=0, parse_dates=['date'])
new_row = {'date': pd.to_datetime('2020-01-01'), '설명': '새해', '요일':'0', 'IsHoliday': 1, '지속휴일수': 1}
holi = holi.append(new_row, ignore_index=True)
holi
```

```python
holi = holi[['date', '설명', 'IsHoliday', '지속휴일수']]
holi.rename(columns={'date': 'real_date'}, inplace=True)
npf = pd.merge(pf_p, holi, on=['real_date'])
npf
```

```python
npf.isnull().sum()
```

```python
final_pf = pd.merge(npf, wt, on=['real_date', 'hour'])
final_pf
```

```python
final_pf.isnull().sum()
```

```python
final_pf.to_csv("final_performance.csv", mode='w')
```

```python
fpf.to_csv("final_performance_v2.csv", mode='w')
```

```python
cpi = pd.read_excel('CPI.xlsx', header=0)
```

```python
cpi_series = cpi.iloc[1]
cpi_dict = cpi_series.to_dict()
del cpi_dict['시도별']
del cpi_dict['품목별']
cpi_dict
```

```python
cpif = pd.DataFrame({'date': list(cpi_dict.keys()), 'cpi': list(cpi_dict.values())})
cpif
```

```python
cpif['year'] = pd.to_datetime(cpif['date']).dt.year
cpif['month'] = pd.to_datetime(cpif['date']).dt.month
cpif
```

```python
del cpif['date']
```

```python
fpf['year'] = pd.to_datetime(fpf['real_date']).dt.year
fpf['month'] = pd.to_datetime(fpf['real_date']).dt.month
fpf.tail()
```

```python
fpfv3 = pd.merge(fpf, cpif, on=['year', 'month'])
fpfv3
```

```python
fpfv3.isnull().sum()
```

```python
fpfv3.to_csv("final_performance_v3.csv", mode='w')
```
